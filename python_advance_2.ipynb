```python
# Inheritance
class parent:
  def __init__(self,name,gender,age):
    self.name=name
    self.gender=gender
    self.age=age
  def welcome(self):
    return f"welcome to python"
  def custom(self):
    return f"{self.name} is a {self.gender} working at the age of {self.age}"
class child(parent):
   pass
p=parent("Radha","Female",35)
print(p.name)
print(p.gender)
print(p.age)
print(p.welcome())
print(p.custom())
ch=child("Chinnu","male",7)
print(ch.name)
print(ch.gender)
print(ch.age)
print(ch.welcome())
print(ch.custom())
```

```python
# Q1):Take a list of integers and return sum of all even numbers
lst=[1,2,3,4,5]
def sum_even_nums(lst):
    even_sum=0
    for i in lst:
      if i%2==0:
        even_sum+=i
    return even_sum
result=sum_even_nums(lst)
print(result)
          # or
def sum_even_numbers(numbers):
    return sum(i for i in numbers if i % 2 == 0)
```

```python
# Q3:write a recursive function in python to compute the numbers in the fibonacci seqence,where the first two numbers
# in the sequence are 0 and 1 and each subsequent number is the sum of the previous two:
def fibonacci(n):
  if n  <=1:
    return n
  else:
    return fibonacci(n-1)+fibonacci(n-2)
print(fibonacci(11))

```

```python
# High order function
  # In python it is a function as arguments or return a function as its result.
  # Map
  # It takes a function and iterable as arguments and apllies the function to each element in the iterable,returning a new iterable.
numbers=[1,2,3,4,5]
def doub_num(num):
  return num*2
doubled_num=list(map(doub_num,numbers))
print(doubled_num)

```

```python
# Q:WAP that uses a map function to convert a tuple of wts in pounds(173,154,112,340,230) to a tuple of wts in kgs.The conversion formula forn pounds to kgs is
# kgs*pounds/2.2.The output should be rounded to 2 decimal places.
pounds=(173,154,112,340,230)
def wt_kgs(i):
  return round(i/2.2 ,2)
wts_in_kgs=tuple(map(wt_kgs,pounds))
print(wts_in_kgs)
# or
pounds=(173,154,112,340,230)
wt_kgs=tuple(map(lambda x:round(x/2.2,2),pounds))
print(wt_kgs)
```
