```python
dic={"name":["radha","alex","arav","abhi","rrr"],
     "dob":["1990-04-10","1991-02-11","1992-03-15","1993-05-23","1994-09-09"],
     "sales":[200,300,400,500,600],"region":["north","east","west","south","soth_east"]}
df=pd.DataFrame(dic)
df
```

```python
df["%_contibution"]=df["sales"]/df["sales"].sum()*100
df["%_contibution"]
```

```python
import matplotlib.pyplot as plt
plt.bar(df["region"],df["sales"])
plt.title("summary of total")
plt.xlabel("region")
plt.ylabel("sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(df["region"],df["%_contibution"])
plt.title("summary of total")
plt.xlabel("region")
plt.ylabel("% contri")
plt.show()
```
