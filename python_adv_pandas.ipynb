```python
# re.search
gene="ATGCAATCCGGAATTCCCGGTAGCTAGCTAGC"
ma=re.search(pattern,gene)
print(ma)
if ma:
  print("match found at index %d : %s"% (ma.start(),ma.group()))
else:
  print("No match")
```

```python
# Finditer
text="ATGCAATCCGGAATTCCCGGTAGCTAGCTAGC"
pattern="[AGCT]"
matches=re.finditer(pattern,text)
print(matches)
for match in matches:
  print("Match:", match.group(), "found at index:", match.start())
```

```python
# span
text="GC@@AATCCG@@$%GAATTCCCGG$#@TAGCTAGCT"
pattern="[^AGCT]"
x=re.search(pattern,text)
print(x.span())
```

```python
# group
text="GC@@AATCCG@@$%GAATTCCCGG$#@TAGCTAGCT"
pattern="[^AGCT]"
x=re.search(pattern,text)
print(x)
print(x.group())
```

```python
# sub
text="GC@@AATCCG@@$%GAATTCCCGG$#@TAGCTAGCT"
print(re.sub("GCGCGC","=ATTACHED",text))
```
